```python
import os
import torch
root_dir = "./ds000030_cleaned_data/img/"  # 改成你自己的路径

# 正确的目标 shape
expected_shape = (96, 96, 96)

# 统计
count_total = 0
count_wrong = 0

# 遍历所有子目录和文件
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith(".pt"):
            file_path = os.path.join(dirpath, filename)
            try:
                data = torch.load(file_path)
                if isinstance(data, torch.Tensor):
                    count_total += 1
                    if data.shape != expected_shape:
                        print(f"[WRONG SHAPE] {file_path} -> {tuple(data.shape)}")
                        count_wrong += 1
                else:
                    print(f"[SKIPPED] Not a tensor: {file_path}")
            except Exception as e:
                print(f"[ERROR] Failed to load {file_path}: {e}")

print(f"\nTotal files checked: {count_total}")
print(f"Files with wrong shape: {count_wrong}")
```

```python
from nilearn import datasets

# 皮层（cortical）atlas
atlas_cort = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-2mm')
print("Cortical atlas:", atlas_cort.maps)

# 亚皮层（subcortical）atlas
atlas_sub = datasets.fetch_atlas_harvard_oxford('sub-maxprob-thr25-2mm')
print("Subcortical atlas:", atlas_sub.maps)

# 全脑（cortical + subcortical）
atlas_whole = datasets.fetch_atlas_harvard_oxford('wholebrain-maxprob-thr25-2mm')
print("Wholebrain atlas:", atlas_whole.maps)
```
